```python
def getStreet(list_id, l):
    try:
        return l.find("p", {"id":"listing-street-" + str(list_id)}).get_text()
    except:
        return "Not Applicable, "
```

```python
def getCity(list_id, l):
    try:
        return l.find("span", {"id":"listing-city-" + str(list_id)}).get_text()
    except:
        return "Not Available"
```

```python
def getState(list_id, l):
    try:
        return l.find("span", {"id":"listing-state-" + str(list_id)}).get_text()
    except:
        return "Not Available"
```

```python
def getZip(list_id, l):
    try:
        return l.find("span", {"id":"listing-zip-" + str(list_id)}).get_text().strip()
    except:
        return "Not Available"
```

```python
def getWeb(list_ld, l):
    try:
        return l.find("a")
    except:
        return "Find Manually"
```

```python
def addAlcoholStoresToDict(lists, city_name):
    list_id = 0
    for l in lists:
        list_id += 1
        if (getCity(list_id, l).startswith(city_name)):
            alc_location_dict["Store Name"].append(l.find("a", {"class" : "business-name"}).get_text())
            alc_location_dict["Address"].append(getStreet(list_id, l) + getCity(list_id, l) + getState(list_id, l))
            alc_location_dict["Website"].append(getWeb(list_id, l))
            alc_location_dict["Phone Number"].append(l.find("span", {"class":"business-phone"}).get_text())
            alc_location_dict["Zip Code"].append(l.find("span", {"id":"listing-zip-" + str(list_id)}).get_text())
```

```python
def addAlcoholStoresToDictZip(lists, zipcode):
    list_id = 0
    for l in lists:
        list_id += 1
        if (getZip(list_id, l).startswith(zipcode)):
            alc_location_dict["Store Name"].append(l.find("a", {"class" : "business-name"}).get_text())
            alc_location_dict["Address"].append(getStreet(list_id, l) + getCity(list_id, l) + getState(list_id, l))
            alc_location_dict["Phone Number"].append(l.find("span", {"class":"business-phone"}).get_text())
            alc_location_dict["Zip Code"].append(l.find("span", {"id":"listing-zip-" + str(list_id)}).get_text())
```

```python
def nextPage(soup):
    try:
        b = True
        linkList = soup.find("ul", {"class" : "pagination pagination-sm pull-right"})
        disabled = linkList.find_all("li", {"class":"disabled"})
        
        if (len(disabled) == 0):
            return True
        for link in disabled:
            if (b == True):
                if (link.find("strong").get_text().startswith("Next")):
                    b = False
                else:
                    b = True
        return b
    except:
        return False
```
